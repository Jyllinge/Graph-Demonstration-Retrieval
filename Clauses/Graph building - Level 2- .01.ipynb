```python
import pickle
import pandas as pd
import numpy as np
import torch
import transformers
import bitsandbytes as bnb
import networkx as nx

random_state = 1
```

```python
def euclid(list1, list2):
    return sum((p-q)**2 for p, q in zip(list1, list2)) ** .5
```

```python
with open('df_policies_embeddings.pkl', 'rb') as f:
    df = pickle.load(f)
df
```

```python
validation_set = ['X_Twitter.txt', 'LinkedIn.txt', 'Zoom.txt', 'Fitbit.txt', 'Western Union.txt', 'Ubisoft.txt']
test_set =  ['Yahoo.txt', 'TikTok.txt', 'MySugr.txt', 'Deliveroo.txt', 'Paypal.txt', 'AdobePhotoshop.txt']
nontrain_set = validation_set + test_set
nontrain_set
```

```python
df_train = df[~df.document_name.isin(nontrain_set)]
df_train = df_train.dropna(subset = ['Level'])
df_train
```

```python
df_train_lv2 = df_train[df_train['Level'] == '2']
df_train_lv2
```

```python
distances = []
for j in range(len(df_train_lv2)):
    newlist = []
    for i in range(len(df_train_lv2)):
        newlist.append(euclid(df_train_lv2.iloc[j]['embeddings'], df_train_lv2.iloc[i]['embeddings']))
    distances.append(newlist)
    print(j)
```

sqrt(316*25) = 88.882

```python
closenodes = []
for el in distances:
    count = 0
    for newel in el:
        if newel <= 45.10:      #set manually, with 63.42 on average the extracted subgraph has (N*25)**0.5 nodes, with 47.49 the nodes are connected to 1% of other nodes, on average
                                #72.87 double nodes                                                                 45.10, .5% of other nodes                                                      
                                #58.47 half nodes                                                                   49.81, 2% of other nodes
            count += 1
    closenodes.append(count)

print(sum(closenodes)/len(closenodes))  #average size of extracted subgraph
print(min(closenodes))                  #min degree
print(max(closenodes))                  #max degree
```

```python
Training_graph = nx.Graph()

for i in range(len(df_train_lv2)):
    print('adding node',i)
    Training_graph.add_node(i)
    for j in range(i):
        if euclid(df_train_lv2.iloc[j]['embeddings'], df_train_lv2.iloc[i]['embeddings']) <= 47.49:      #così in media ogni nodo è collegato all'1% degli altri nodi, in media
            Training_graph.add_edge(i,j)

pickle.dump(Training_graph, open('lv2graph_Level_.01.pickle', 'wb'))
nx.draw_networkx(Training_graph, with_labels = False)
```
